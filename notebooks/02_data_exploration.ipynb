```python
# Cell 1: Load Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)

X = mnist.data.to_numpy()
y = mnist.target.astype(int)
# Cell 2: Data Shape & Info
print(f"Total samples: {X.shape[0]}")
print(f"Features per sample: {X.shape[1]} (8x8 pixels flattened)")
print(f"Pixel value range: [{X.min()}, {X.max()}]")
print(f"Classes: {np.unique(y)}")

# Cell 3: Pixel Intensity Distribution per Digit
fig, axes = plt.subplots(2, 5, figsize=(15, 8))
for digit in range(10):
    ax = axes[digit // 5, digit % 5]
    digit_pixels = X[y == digit]
    ax.hist(digit_pixels.flatten(), bins=30, alpha=0.7, color='steelblue')
    ax.set_title(f"Pixel Intensity Distribution - Digit {digit}")
    ax.set_xlabel("Pixel Value")
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.savefig('../data/pixel_intensity_distributions.png', dpi=100, bbox_inches='tight')
plt.show()

# Cell 4: Outlier Detection & Missing Values
print(f"Missing values: {np.isnan(X).sum()}")
print(f"Infinite values: {np.isinf(X).sum()}")
print(f"Zeros: {(X == 0).sum()}")
print(f"Max values: {(X == X.max()).sum()}")
```
