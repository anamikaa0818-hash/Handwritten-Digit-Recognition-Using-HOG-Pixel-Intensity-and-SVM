HANDWRITTEN DIGIT RECOGNITION USING CLASSICAL ML

Problem Statement:
- Classify handwritten digits (0-9) from MNIST dataset
- Use classical ML (not neural networks)
- Feature extraction: HOG, Pixel Intensity, Zernike Moments
- Compare: SVM, KNN, Logistic Regression, Random Forest

Why Classical ML?
- Interpretability: Understand which features matter
- Computational efficiency: No GPU required
- Industry relevance: Many production systems use traditional ML
- Understanding fundamentals before deep learning

Dataset: MNIST
- 70,000 images (60,000 train, 10,000 test)
- 28x28 grayscale pixels
- 10 classes (digits 0-9)
- Balanced dataset


```python

# Cell 2: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import cv2
from skimage import feature, transform
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

```

Cell 3: Load MNIST Dataset (via sklearn digits)
Cell 3: Load FULL MNIST Dataset (28x28)


```python

from sklearn.datasets import fetch_openml

print("Downloading MNIST dataset...")

mnist = fetch_openml('mnist_784', version=1, as_frame=False)

X_full = mnist.data          # Shape: (70000, 784)
y_full = mnist.target.astype(int)

print(f"Dataset shape: {X_full.shape}")
print(f"Image size: 28x28")
print(f"Number of classes: {len(np.unique(y_full))}")

print("\nClass distribution:")
print(pd.Series(y_full).value_counts().sort_index())
# Cell 4: Visualize Sample Digits
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_full[i].reshape(28, 28), cmap='gray')
    ax.set_title(f"Digit: {y_full[i]}")
    ax.axis('off')
plt.suptitle("Sample Handwritten Digits from Dataset")
plt.tight_layout()
plt.savefig('sample_digits.png', dpi=100, bbox_inches='tight')
plt.show()

# Cell 5: Analyze Class Distribution
plt.figure(figsize=(10, 5))
sns.countplot(x=y_full, palette='viridis')
plt.xlabel("Digit Class")
plt.ylabel("Frequency")
plt.title("Class Distribution in Dataset")
plt.tight_layout()
plt.savefig('../data/class_distribution.png', dpi=100, bbox_inches='tight')
plt.show()
```
